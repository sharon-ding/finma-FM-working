```python
import pandas as pd

# create dictionary with 4 entries
my_data = {                     
    'Date': pd.date_range(start='2023-01-01', periods=5, freq='D'),       # date_range: calendar building feature in pandas
    'AAPL': [150.75, 153.30, 149.50, 155.00, 157.25],                     # creates a series of 5 dates
    'GOOGL': [2800.50, 2820.00, 2790.00, 2850.00, 2900.00],
    'MSFT': [299.00, 305.00, 300.00, 310.00, 315.00]
}
df = pd.DataFrame(my_data)     # create a new data frame, and load the my_data dictionary into it
print(df)
```

```python
# line chart of sample prices
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use("seaborn-v0_8")

dates = pd.date_range(start="2023-01-01", periods=5, freq="D")
prices = [150.75, 153.30, 149.50, 155.00, 157.25]

plt.figure(figsize=(8, 4))
plt.plot(dates, prices, marker="o", label="AAPL close")
plt.title("Sample Price History")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
