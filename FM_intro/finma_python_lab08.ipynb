### Exercise 2: Return Statistics (NumPy)

Given daily returns for a stock, calculate:
1. Average daily return
2. Volatility (standard deviation)
3. Annualized return (assume 252 trading days)
4. Annualized volatility
5. Sharpe ratio (assume risk-free rate = 2%)

**Formula:** Sharpe Ratio = (Annualized Return - Risk Free Rate) / Annualized Volatility

**Data:** Use returns array: `[1.2, -0.5, 0.8, 1.5, -0.3, 0.9, 1.1, -0.7, 1.3, 0.6]`

````python
# Exercise 2: Return Statistics
# Write your code here
import numpy as np
daily_returns = np.array([1.2, -0.5, 0.8, 1.5, -0.3, 0.9, 1.1, -0.7, 1.3, 0.6])
                         
# average daily return
average_daily_return = daily_returns.sum()/len(daily_returns)

# volatility
standard_deviation = daily_returns.std()

# annualised return
annualised_return = (1+average_daily_return)**252 - 1

# annualised volatility
annualised_volatility = np.std(daily_returns)*np.sqrt(252)

# sharpe ratio
sharpe_ratio = (annualised_return-0.02)/annualised_volatility

print("Average daily return:", average_daily_return)
print("Volatility:", standard_deviation)
print("Annualised return", annualised_return)
print("Annualised volatility:", annualised_volatility)
print("Sharpe ratio:", sharpe_ratio)
````

### Exercise 5: Transaction Analysis (Pandas)

Read transactions.csv and:

2. Calculate total commission paid
3. Calculate average buy price and average sell price for each symbol
4. Determine current holdings (shares bought - shares sold)
5. Export a summary to CSV with columns: symbol, shares_held, avg_buy_price, total_commission

````python
# Exercise 5: Transaction Analysis
# Write your code here
import pandas as pd

transactions = pd.read_csv('transactions.csv')
print(transactions)

print(transactions['commission'].sum())

print('\n3. Average prices:')
avg_prices = transactions.groupby(['symbol', 'action']).agg({
    'price': 'mean',
    'shares': 'sum'
})
print(avg_prices)

print('\n4. Current holdings:')
buys = transactions[transactions['action'] == 'BUY'].groupby('symbol')['shares'].sum()
sells = transactions[transactions['action'] == 'BUY'].groupby('symbol')['shares'].sum()
holdings = buys.subtract(sells, fill_value=0)
print(holdings)

# export summary
avg_buy_prices = transactions[transactions['action'] == 'BUY'].groupby('symbol')['price'].mean()
commission_by_symbol = transactions.groupby('symbol')['commission'].sum() 

export_df = pd.DataFrame({
    'symbol': holdings.index,
    'shares_held': holdings.values, 
    'avg_buy_price': avg_buy_prices[holdings.index].values,
    'total_commission': commission_by_symbol[holdings.index].values    
})

export_df.to_csv('holdings_summary.csv', index=False)
print('\n5.Summary exported to holdings_summary.csv')
print(export_df)
````

### Exercise 8: Portfolio Performance Report (Combined)

Create a comprehensive portfolio performance report:

1. Read transactions.csv to determine current holdings
2. Read stock_prices_timeseries.csv to get latest prices
3. Merge company_info.csv to get sector information
4. Calculate for each holding:
   - Current value
   - Cost basis
   - Unrealized gain/loss
   - Return percentage
5. Calculate portfolio-level metrics:
   - Total value
   - Total cost
   - Total gain/loss
   - Overall return %
   - Allocation by sector
6. Export to CSV with all details

**Hint:** Use the last price in the time series as the current price

````python
# Exercise 8: Portfolio Performance Report
# Write your code here

import pandas as pd
import numpy as np

with open('transactions.csv', 'r') as file:
    transactions = pd.read_csv('transactions.csv')
    transactions_com = pd.DataFrame(transactions)

with open('stock_prices_timeseries.csv', 'r') as file:
    sp = pd.read_csv('stock_prices_timeseries.csv')
    sp_com = pd.DataFrame(sp)
with open( 'company_info.csv', 'r') as file:
    ci = pd.read_csv('company_info.csv')
    ci_com = pd.DataFrame(ci)


combined_df = pd.concat([sp_com, ci_com, transactions_com], ignore_index=True)
#merged_df = pd.merge(sp_com, ci_com, on='symbol')
print (combined_df)

combined_df['CurrentValue'] = combined_df['shares'] * combined_df['price']
print ("Current value: ", combined_df['CurrentValue'])

combined_df['cost_basis'] = (combined_df['CurrentValue']-combined_df['commission'])/combined_df['shares']
print ("Cost value: ", combined_df['cost_basis'])

combined_df['unrealisedgl'] = combined_df['CurrentValue']-combined_df['cost_basis']
print ("Unrealised gains/loss", combined_df['unrealisedgl'])

combined_df['return_percentage'] = ( combined_df['unrealisedgl'] / combined_df['cost_basis']) * 100
print ("Return percentage: ", combined_df['return_percentage'])

total_value = combined_df['CurrentValue'].sum()
print ("Total value: ", total_value)

total_cost = combined_df['cost_basis'].sum()
print("Total cost: ", total_cost)

total_profit_loss = combined_df['unrealisedgl'].sum()
print("Total profit/loss: ", total_profit_loss)

overall_return_pct = ((combined_df['CurrentValue'].sum() - combined_df['cost_basis'].sum())/ combined_df['cost_basis'].sum()) * 100
print("Overall return %: ", overall_return_pct)

sector_allocation = (combined_df.groupby('sector')['CurrentValue'].sum()/ total_value* 100)
print("Sector allocation: ", sector_allocation)



combined_df.to_csv('combined_df.csv')

print("\nFiles exported successfully to the output folder.")
````
