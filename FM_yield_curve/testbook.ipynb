```python
from datetime import datetime, timedelta
import pandas as pd

# Instantiate a zero curve
zero_curve = {
    'dates': [datetime(2024, 1, 1), datetime(2024, 6, 1), datetime(2025, 1, 1)],
    'rates': [0.02, 0.025, 0.03]
}

# Create two bank bills
bill1 = {
    'name': 'Bill 1',
    'face_value': 100,
    'maturity_date': datetime(2024, 3, 1),
    'issue_date': datetime(2024, 1, 1),
    'discount_rate': 0.02,
    'cashflows': []
}

bill2 = {
    'name': 'Bill 2',
    'face_value': 100,
    'maturity_date': datetime(2024, 9, 1),
    'issue_date': datetime(2024, 1, 1),
    'discount_rate': 0.025,
    'cashflows': []
}

# Create two bonds
bond1 = {
    'name': 'Bond 1',
    'face_value': 100,
    'coupon_rate': 0.05,
    'maturity_date': datetime(2026, 1, 1),
    'issue_date': datetime(2024, 1, 1),
    'cashflows': []
}

bond2 = {
    'name': 'Bond 2',
    'face_value': 100,
    'coupon_rate': 0.04,
    'maturity_date': datetime(2025, 6, 1),
    'issue_date': datetime(2024, 1, 1),
    'cashflows': []
}

# Set cash flows for instruments
bill1['cashflows'] = [(bill1['maturity_date'], bill1['face_value'])]
bill2['cashflows'] = [(bill2['maturity_date'], bill2['face_value'])]

# Bond 1: semi-annual coupons
bond1_dates = pd.date_range(start=datetime(2024, 7, 1), end=bond1['maturity_date'], freq='6MS')
bond1['cashflows'] = [(d, bond1['face_value'] * bond1['coupon_rate'] / 2) for d in bond1_dates[:-1]]
bond1['cashflows'].append((bond1['maturity_date'], bond1['face_value'] * (1 + bond1['coupon_rate'] / 2)))

# Bond 2: semi-annual coupons
bond2_dates = pd.date_range(start=datetime(2024, 7, 1), end=bond2['maturity_date'], freq='6MS')
bond2['cashflows'] = [(d, bond2['face_value'] * bond2['coupon_rate'] / 2) for d in bond2_dates[:-1]]
bond2['cashflows'].append((bond2['maturity_date'], bond2['face_value'] * (1 + bond2['coupon_rate'] / 2)))

# Create portfolio
portfolio = {
    'name': 'Portfolio',
    'instruments': [bill1, bill2, bond1, bond2]
}

# Print PV of cash flows for each instrument
today = datetime(2024, 1, 1)
for instrument in portfolio['instruments']:
    pv = 0
    for date, amount in instrument['cashflows']:
        years = (date - today).days / 365
        # Using simple linear interpolation for discount rate
        discount_rate = 0.025  # Simplified assumption
        pv += amount / ((1 + discount_rate) ** years)
    print(f"{instrument['name']}: PV = ${pv:.2f}")

# Total portfolio PV
portfolio_pv = sum(
    sum(
        amount / ((1 + 0.025) ** ((date - today).days / 365))
        for date, amount in instrument['cashflows']
    )
    for instrument in portfolio['instruments']
)
print(f"\nTotal Portfolio PV: ${portfolio_pv:.2f}")
```
